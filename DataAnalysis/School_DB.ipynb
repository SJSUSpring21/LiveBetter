```python
import pandas as pd
import numpy as np
```

```python
df_colleges = pd.read_csv('School Data/Colleges_and_Universities.csv')
df_public = pd.read_csv('School Data/Public_Schools.csv')
df_private = pd.read_csv('School Data/Private_Schools.csv')
```

```python
df_public.head()
```

```python
df_colleges_new = df_colleges[['NAME','LATITUDE','LONGITUDE']]
df_colleges_new.columns=['NAME','lat','lng']
df_colleges_new.to_csv('School Data/Colleges.csv',index=False)
df_colleges_new.head()
```

```python
df_public_new = df_public[['NAME','LATITUDE','LONGITUDE']]
df_public_new.columns=['NAME','lat','lng']
df_public_new.to_csv('School Data/Public.csv',index=False)
df_public_new.head()
```

```python
df_private_new = df_private[['NAME','LATITUDE','LONGITUDE']]
df_private_new.columns=['NAME','lat','lng']
df_private_new.to_csv('School Data/Private.csv',index=False)
df_private_new.head()
```

```python
df_schools = pd.concat([df_colleges_new,df_public_new,df_private_new])
df_schools.to_csv('School Data/Schools.csv',index=False)
df_schools.head()
```
